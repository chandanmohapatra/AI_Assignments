Create a dictionary that stores student names as keys and marks as values, then find the student with the highest marks.

```python
student = {"chandan":{"marks":84, "Course":"Python"},
           "Sarika":{"marks": 60, "course":"HR"}}
top_student = max(student, key=lambda x: student[x]['marks'])
print(top_student)
top_marks = student[top_student]['marks']
print(top_marks)
#
```

Implement a while loop to print the Fibonacci series up to a given number.

```python
n = int(input("enter a number"))
a,b = 0,1
next = b
count=1
print(a)
while count<=n:
  print(next)
  next = a+b
  a,b = b,next
  count=count+1
```
